```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC','^IXIC','^GDAXI']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data/ ind_data.iloc[0] * 100).plot(figsize=(15,6))
plt.show()
```

```python
# Around the millimum we observe a general rise of the indices value.This was because in the period between 1995 and 2001 there was
# a dot.com bubble and the stock markets in dustrialized countries like USA UK and Germany were boosted by a significant rise in the value
# of Internet companies being composed primarily of tech companies. Nasdaq logically depicts best. Of course, the global economic and financial crises of 2008 as well as the sovereign debt crisis in 2011
# reflect the picture accordingly. But in the long term observing the resultes from 1997 until today we can say that overall indices 
# performed very well with serval occasional drops. Nasdaq scores highest in certain periods and its performance is connected to slightly 
# sharper variations at times. The German market is the one that demonstrates real persistence and stability.
```

```python
ind_returns = (ind_data / ind_data.shift(1)) -1 
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 250   # namely annual returns
annual_ind_returns
```
